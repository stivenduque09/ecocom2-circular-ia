```python
# -*- coding: utf-8 -*-
"""
EcoCom2 Circular IA — Módulo de detección mejorado
Reemplaza la lógica de YOLO del archivo original con:
  - Umbral de confianza apropiado (conf=0.40)
  - IoU para evitar detecciones duplicadas del mismo objeto
  - Validación de imagen antes de procesar
  - Mapa de materiales extendido (35+ objetos)
  - Deduplicación por clase para contar correctamente
  - Función de análisis reutilizable (lista para chatbot)
"""

!pip install ultralytics
from PIL import Image
from ultralytics import YOLO
import os

# ─────────────────────────────────────────────
# CARGAR MODELO (una sola vez al inicio)
# ─────────────────────────────────────────────

print("🤖 Cargando modelo de IA...")
modelo = YOLO("yolov8m.pt")
print("✅ Modelo listo.")


# ─────────────────────────────────────────────
# PARÁMETROS DE DETECCIÓN
# ─────────────────────────────────────────────

CONFIANZA_MINIMA = 0.40   # Antes era 0.1 — muy permisivo, generaba falsos positivos
IOU_THRESHOLD    = 0.50   # Evita contar el mismo objeto dos veces (non-maximum suppression)


# ─────────────────────────────────────────────
# OBJETOS A IGNORAR (no son residuos)
# ─────────────────────────────────────────────

IGNORAR = {
    "person", "car", "bus", "truck", "motorcycle", "bicycle",
    "dog", "cat", "bird", "horse", "sheep", "cow", "elephant",
    "bear", "zebra", "giraffe", "traffic light", "fire hydrant",
    "stop sign", "parking meter", "bench", "potted plant",
    "oven", "refrigerator", "toilet", "sink", "bed",
    "dining table", "tv", "remote", "clock", "vase",
    "scissors", "hair drier", "toothbrush", "sports ball",
    "kite", "baseball bat", "baseball glove", "skateboard",
    "surfboard", "tennis racket", "fire extinguisher",
    "tie", "scarf", "frisbee", "skis", "snowboard",
    "golf club", "umbrella", "handbag",
}


# ─────────────────────────────────────────────
# MAPA DE MATERIALES (extendido de 16 a 35+ objetos)
# ─────────────────────────────────────────────

MATERIALES = {

    # ── PLÁSTICOS ──────────────────────────────
    "bottle":       {"es": "Botella plástica",  "material": "Plástico",  "peso_kg": 0.05},
    "cup":          {"es": "Vaso desechable",   "material": "Plástico",  "peso_kg": 0.03},
    "bowl":         {"es": "Recipiente",        "material": "Plástico",  "peso_kg": 0.08},
    "chair":        {"es": "Silla",             "material": "Plástico",  "peso_kg": 2.0},

    # ── PAPEL Y CARTÓN ─────────────────────────
    "book":         {"es": "Libro",             "material": "Papel",     "peso_kg": 0.20},
    "box":          {"es": "Caja de cartón",    "material": "Cartón",    "peso_kg": 0.20},
    "notebook":     {"es": "Cuaderno",          "material": "Papel",     "peso_kg": 0.18},

    # ── VIDRIO ─────────────────────────────────
    "wine glass":   {"es": "Copa/frasco vidrio","material": "Vidrio",    "peso_kg": 0.30},
    "vase":         {"es": "Vaso de vidrio",    "material": "Vidrio",    "peso_kg": 0.25},

    # ── METALES ────────────────────────────────
    "can":          {"es": "Lata de aluminio",  "material": "Aluminio",  "peso_kg": 0.02},
    "fork":         {"es": "Tenedor metálico",  "material": "Metal",     "peso_kg": 0.05},
    "knife":        {"es": "Cuchillo metálico", "material": "Metal",     "peso_kg": 0.06},
    "spoon":        {"es": "Cuchara metálica",  "material": "Metal",     "peso_kg": 0.04},

    # ── ELECTRÓNICOS ───────────────────────────
    "tv":           {"es": "Televisor",         "material": "Electrónico","peso_kg": 6.0},
    "laptop":       {"es": "Portátil",          "material": "Electrónico","peso_kg": 2.0},
    "cell phone":   {"es": "Celular",           "material": "Electrónico","peso_kg": 0.20},
    "keyboard":     {"es": "Teclado",           "material": "Electrónico","peso_kg": 0.80},
    "mouse":        {"es": "Mouse",             "material": "Electrónico","peso_kg": 0.10},
    "remote":       {"es": "Control remoto",    "material": "Electrónico","peso_kg": 0.09},

    # ── ORGÁNICOS ──────────────────────────────
    "banana":       {"es": "Residuo orgánico",  "material": "Orgánico",  "peso_kg": 0.12},
    "apple":        {"es": "Residuo orgánico",  "material": "Orgánico",  "peso_kg": 0.18},
    "orange":       {"es": "Residuo orgánico",  "material": "Orgánico",  "peso_kg": 0.15},
    "sandwich":     {"es": "Residuo orgánico",  "material": "Orgánico",  "peso_kg": 0.10},
    "pizza":        {"es": "Residuo orgánico",  "material": "Orgánico",  "peso_kg": 0.30},
    "hot dog":      {"es": "Residuo orgánico",  "material": "Orgánico",  "peso_kg": 0.08},

    # ── ROPA / TEXTILES ────────────────────────
    "backpack":     {"es": "Mochila",           "material": "Textil",    "peso_kg": 1.0},
    "suitcase":     {"es": "Maleta",            "material": "Material mixto","peso_kg": 3.0},
    "handbag":      {"es": "Bolso",             "material": "Textil",    "peso_kg": 0.5},

    # ── MUEBLES / GRANDES ──────────────────────
    "couch":        {"es": "Sofá",              "material": "Material mixto","peso_kg": 15.0},
    "bed":          {"es": "Colchón/cama",      "material": "Material mixto","peso_kg": 20.0},

    # ── OTROS APROVECHABLES ────────────────────
    "umbrella":     {"es": "Paraguas",          "material": "Material mixto","peso_kg": 0.4},
    "clock":        {"es": "Reloj dañado",      "material": "Electrónico","peso_kg": 0.3},
    "scissors":     {"es": "Tijeras",           "material": "Metal",     "peso_kg": 0.08},
    "toothbrush":   {"es": "Cepillo dental",    "material": "Plástico",  "peso_kg": 0.02},
    "hair drier":   {"es": "Secadora pelo",     "material": "Electrónico","peso_kg": 0.5},
}


# ─────────────────────────────────────────────
# VALIDAR IMAGEN
# ─────────────────────────────────────────────

def validar_imagen(ruta_archivo):
    """
    Verifica que el archivo sea una imagen válida antes de procesarla.
    Retorna (True, None) si es válida, o (False, "mensaje de error") si no.
    """
    if not os.path.exists(ruta_archivo):
        return False, "❌ El archivo no existe."

    tamanio_bytes = os.path.getsize(ruta_archivo)
    if tamanio_bytes == 0:
        return False, "❌ El archivo está vacío."
    if tamanio_bytes > 15 * 1024 * 1024:  # 15 MB máximo
        return False, "❌ La imagen es demasiado grande (máx. 15 MB)."

    try:
        with Image.open(ruta_archivo) as img:
            img.verify()  # Verifica que no esté corrupta
    except Exception:
        return False, "❌ El archivo no es una imagen válida o está dañado."

    return True, None


# ─────────────────────────────────────────────
# FUNCIÓN PRINCIPAL DE ANÁLISIS
# ─────────────────────────────────────────────

def analizar_imagen(ruta_archivo):
    """
    Analiza una imagen con YOLO y retorna los materiales reciclables detectados.

    Retorna un diccionario con:
      - "detectados": lista de materiales encontrados
      - "peso_total_kg": suma de pesos estimados
      - "clasificacion_zona": "bajo", "medio", "alto" o "sin_evidencia"
      - "error": None si todo fue bien, o string con el mensaje de error
    """

    # 1. Validar imagen antes de procesar
    valida, error_msg = validar_imagen(ruta_archivo)
    if not valida:
        return {"detectados": [], "peso_total_kg": 0, "clasificacion_zona": "sin_evidencia", "error": error_msg}

    # 2. Correr YOLO con umbrales correctos
    try:
        resultados = modelo(
            ruta_archivo,
            conf=CONFIANZA_MINIMA,   # 0.40 — certeza mínima para reportar
            iou=IOU_THRESHOLD,       # 0.50 — evita duplicados del mismo objeto
            verbose=False
        )
    except Exception as e:
        return {"detectados": [], "peso_total_kg": 0, "clasificacion_zona": "sin_evidencia", "error": f"❌ Error al analizar: {e}"}

    # 3. Procesar resultados con deduplicación por clase
    clases_vistas = set()     # Para no contar el mismo tipo de objeto dos veces
    detectados = []
    peso_total = 0.0

    for r in resultados:
        for caja in r.boxes:
            clase_id    = int(caja.cls[0])
            nombre_obj  = modelo.names[clase_id].lower()
            confianza   = float(caja.conf[0])

            # Ignorar objetos que no son residuos
            if nombre_obj in IGNORAR:
                continue

            # Buscar en el mapa de materiales
            info = MATERIALES.get(nombre_obj)
            if info is None:
                continue

            # MEJORA CLAVE: deduplicar por clase
            # Si ya detectamos una "botella", no la contamos dos veces
            if nombre_obj in clases_vistas:
                continue
            clases_vistas.add(nombre_obj)

            detectados.append({
                "nombre":     info["es"],
                "material":   info["material"],
                "peso_kg":    info["peso_kg"],
                "confianza":  round(confianza * 100, 1),  # en porcentaje
            })
            peso_total += info["peso_kg"]

    # 4. Clasificar la zona según cantidad de residuos detectados
    cantidad = len(detectados)
    if cantidad >= 6:
        zona = "alto"
    elif cantidad >= 3:
        zona = "medio"
    elif cantidad >= 1:
        zona = "bajo"
    else:
        zona = "sin_evidencia"

    return {
        "detectados":        detectados,
        "peso_total_kg":     round(peso_total, 2),
        "clasificacion_zona": zona,
        "error":             None,
    }


# ─────────────────────────────────────────────
# FUNCIÓN DE PRESENTACIÓN (para el menú de Colab)
# ─────────────────────────────────────────────

ZONA_EMOJIS = {
    "alto":          "🔴 Punto crítico alto",
    "medio":         "🟡 Punto crítico medio",
    "bajo":          "🟢 Punto crítico bajo",
    "sin_evidencia": "⚪ Sin evidencia suficiente",
}

def mostrar_resultado(resultado, ubicacion, direccion, latitud, longitud):
    """
    Imprime el resumen del análisis en el menú de consola de Colab.
    Separa presentación de lógica para que el chatbot pueda usar analizar_imagen()
    directamente sin llamar a esta función.
    """
    if resultado["error"]:
        print(resultado["error"])
        return

    detectados = resultado["detectados"]

    if not detectados:
        print("\n❌ No se detectaron materiales reciclables en esta imagen.")
        print("⚠️  El reporte no fue validado.")
        return

    print("\n✅ Materiales detectados:")
    for item in detectados:
        print(f"  • {item['nombre']} ({item['material']}) — conf: {item['confianza']}%  ⚖️ {item['peso_kg']} kg")

    zona_texto = ZONA_EMOJIS[resultado["clasificacion_zona"]]

    print("\n--- Resumen del reporte ---")
    print(f"📍 Barrio:          {ubicacion}")
    print(f"📌 Referencia:      {direccion}")
    if latitud != "N/A":
        print(f"🌍 GPS:             {latitud:.6f}, {longitud:.6f}")
    else:
        print("🌍 GPS:             No disponible")
    print(f"♻️  Materiales:      {', '.join(set(i['material'] for i in detectados))}")
    print(f"⚖️  Peso total:      {resultado['peso_total_kg']} kg")
    print(f"🚦 Clasificación:   {zona_texto}")
    print("\n🤖 EcoCom2 IA finalizó el análisis.")
    print("✅ Reporte validado.\n")


# ─────────────────────────────────────────────
# EJEMPLO DE USO DIRECTO EN COLAB
# ─────────────────────────────────────────────
#
# from google.colab import files
# archivo = files.upload()
# nombre = list(archivo.keys())[0]
#
# resultado = analizar_imagen(nombre)
# mostrar_resultado(resultado, "Andalucía", "Frente al parque", 6.2701, -75.5625)
#
# ─────────────────────────────────────────────
# PARA EL CHATBOT (WhatsApp/Telegram):
# ─────────────────────────────────────────────
#
# resultado = analizar_imagen("/tmp/foto_usuario.jpg")
# if resultado["error"]:
#     enviar_mensaje(resultado["error"])
# else:
#     zona = ZONA_EMOJIS[resultado["clasificacion_zona"]]
#     respuesta = f"Detecté {len(resultado['detectados'])} residuo(s).\n"
#     respuesta += f"Zona: {zona}\n"
#     respuesta += f"Peso aprox: {resultado['peso_total_kg']} kg"
#     enviar_mensaje(respuesta)
```
